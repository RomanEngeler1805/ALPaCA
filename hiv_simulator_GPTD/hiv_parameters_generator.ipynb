```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
perturb_params = ('p_lambda1','p_lambda2','p_k1','p_k2','p_f', \
            'p_m1','p_m2','p_lambdaE','p_bE','p_Kb','p_d_E','p_Kd')
N_samples = 200


parameter_list = []

for i in range(N_samples):
    new_dict = {key: value for (key, value) in zip(perturb_params, np.random.normal(scale=0.1, size=12))}
    parameter_list.append(new_dict)
```

```python
print(len(parameter_list))
idx = np.arange(150, dtype=np.int32)
print(idx[:2])
newlist = [parameter_list[i] for i in range(len(parameter_list))if i <100]
print(len(newlist))
```

```python
parr = np.zeros([N_samples, 12])

for i in range(N_samples):
    pp = parameter_list[i]
    
    for j, key in enumerate(pp):
        parr[i,j] = pp[key]
```

```python
for j in range(12):
    plt.figure()
    plt.hist(parr[:,j])
    plt.show()
```

```python
# dumping file

with open('./'+'hiv'+'_generated_hidden_params','wb') as f:
    pickle.dump(parameter_list, f)
```
